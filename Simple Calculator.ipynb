# Building a Simple Calculator

```python
# Create fist number
num_1 = int(input("Choose the first number: "))

# Create second number
num_2 = int(input("Choose the second number: "))

# Choose the operators (Multiplication (*), Division (/), Addition (+), Subtraction (-))
op_sign = input("Choose the sign: ")

# Do the calculation:

if op_sign == "+":
    print(num_1 + num_2)
elif op_sign == "-":
    print(num_1 - num_2)
elif op_sign == "/":
    print(num_1 / num_2)
elif op_sign == "*":
    print(num_1 * num_2)
else:
    print("Invalid sign")
```
